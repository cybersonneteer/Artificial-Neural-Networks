```python

import numpy as np
import matplotlib.pyplot as plt

# 1. Define the  function
def sigmoid_func(v):
    return np.array((1./(1+np.exp(-a*v))), dtype=float)

# 2. Test with a scalar input
v = 10.5
a = 0.5
y = sigmoid_func(v)
print("y =", y)

# 3. Define a range of input values
v_range = np.arange(-10, 10.1, 0.1)

# 4. Compute the output over the range
y_range = sigmoid_func(v_range)

# 5. Plot the result
plt.figure()
plt.plot(v_range, y_range, linewidth=2)
plt.grid(True)
plt.xlabel('Input v')
plt.ylabel('Output y')
plt.title('Threshold (Hard Limit) Activation Function')
plt.ylim([-0.5, 1.5])
plt.axvline(x=0, color='r', linestyle='--', label='Threshold = 0')
plt.legend()
plt.show()
```
